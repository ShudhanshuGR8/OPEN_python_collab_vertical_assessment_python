Q1) Is there any relationship present between Year Built and Price? If yes, Please explain.

```python
# your code here
sales = pd.read_csv('sample_data/kc_house_data.csv')

a = sales['yr_built'].value_counts()
sales['yr_built_count'] = a
sales_dropna = sales.dropna()
s_d = sales_dropna[['price', 'yr_built']]
v = s_d.sort_values(by=['yr_built'])
print(v)

v.plot(kind="line", figsize=(20,10))
```

As the above line chart is evident, the relationship between year_built and price is very complex and there will be very less effect of each of these features on eachselves. Also, we can analyse the data by seeing that even in the earliest of the years such as 1900 has hgher rate than the latest of the years such as 2013.

Q2 ) Add 1.8% TAX on every sale Price within Zipcode 98103
